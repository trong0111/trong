```python
#Bai3
c=str(input("Nhap vao 2 ki tu: "))
if len(c) != 1:
    print("Khoi dong lai code va chi nhap vao 1 ki tu thoi")
else:
    if 'a' <= c <= 'z':
        print("Ki tu hoa tuong ung la: ", c.upper())
    if 'A' <= c <= 'Z':
        print("Ki tu thuong tuong ung la: ", c.lower())
```

```python
#Bai4
c = input("Nhập một ký tự: ")
if len(c) != 1:
    print("Vui lòng chỉ nhập 1 ký tự!")
else:
    if c.isalpha():
        print(f"{c} là kí tự alphabet")
    else:
        print(f"{c} không phải là kí tự alphabet")
```

```python
#Bai6
import math
a, b, c = map(float, input("Nhap vao 3 so: ").split())
if a+b>c and b+c>a and a+c>b:
    p=(a+b+c)/2
    s=math.sqrt(p*(p-a)*(p-b)*(p-c))
    print(s)
else:
    print("Khong phai 3 canh cua tam giac")
```
